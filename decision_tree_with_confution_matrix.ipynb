```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
data=pd.read_csv("/content/StudentPerformanceFactors.csv")
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.drop('Distance_from_Home', axis=1)
```

```python
data
```

```python
data = data.drop('Teacher_Quality', axis=1)
```

```python
data = data.drop('Parental_Education_Level', axis=1)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

data['Access_to_Resources'] = le.fit_transform(data['Access_to_Resources'])
data['Parental_Involvement'] = le.fit_transform(data['Parental_Involvement'])
data['Family_Income'] = le.fit_transform(data['Family_Income'])
```

```python
data
```

```python
x=data[["Access_to_Resources","Parental_Involvement","Family_Income"]]
```

```python
y=data["Extracurricular_Activities"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dt = DecisionTreeClassifier(criterion="entropy", random_state=42)
```

```python
dt.fit(x_train, y_train)

y_pred = dt.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import precision_score, recall_score,f1_score
```

```python
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
recall
```
